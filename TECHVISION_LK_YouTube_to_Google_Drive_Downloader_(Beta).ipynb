# YouTube video to Google Drive Download and Drive mount session 
>Click on the run cell button and wait for the end and then it ask to mount. Please click on **Connect to Google Drive**. To stream files we need to mount Google Drive. 

```python
from ipywidgets import IntProgress
from IPython.display import display
import time

!pip install pytube

#@title ##TECHVISION LK YouTube to Google Drive Downloader (Beta)
#@markdown ---

#@markdown Please enter your YouTube URL here and Choose the video resolution from bellow 
URL = ""#@param {type:"string"}


from pytube import YouTube
video = YouTube(URL)
video.title

Resolution = "360p"#@param ["144p", "240p", "360p", "480p", "720p", "1080p"]

download = video.streams.filter(res=Resolution)
download

from google.colab import drive
drive.mount('/content/drive')

download.first().download('drive/My Drive/')

max_count = 100

f = IntProgress(min=0, max=max_count) # instantiate the bar
display(f) # display the bar

count = 0
while count <= max_count:
    f.value += 1 # signal to increment the progress bar
    time.sleep(.1)
    count += 1

print(video.title, "Video downloaded. Thanks you for using TECHVISION LK YouTube To Google Drive Downloader !")
```
